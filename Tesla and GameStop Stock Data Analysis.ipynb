```python
import yfinance as yf
import pandas as pd
tesla_data = yf.download('TSLA', period='5y')
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

tables = soup.find_all("table")

for i, table in enumerate(tables):
    print(f"Table {i}")
    print(pd.read_html(str(table))[0].head(), "\n")
tesla_revenue = pd.read_html(str(tables[1]))[0]  
tesla_revenue.columns = ["Date", "Revenue"]
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].replace({'\$':'', ',':''}, regex=True)
tesla_revenue.dropna(inplace=True)
tesla_revenue.tail()
```

```python

gme_data = yf.download('GME', period='5y')

gme_data.reset_index(inplace=True)

gme_data.head()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")
tables = soup.find_all("table")
for i, table in enumerate(tables):
    print(f"Table {i}")
    print(pd.read_html(str(table))[0].head(), "\n")
gme_revenue = pd.read_html(str(tables[1]))[0] 
gme_revenue.columns = ["Date", "Revenue"]
gme_revenue["Revenue"] = gme_revenue["Revenue"].replace({'\$':'', ',':''}, regex=True)
gme_revenue.dropna(inplace=True)
gme_revenue.tail()
```

```python
import matplotlib.pyplot as plt
def make_graph(stock_data, title):
    plt.figure(figsize=(10,5))
    plt.plot(stock_data['Date'], stock_data['Close'], label='Closing Price')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Closing Price (USD)')
    plt.legend()
    plt.show()
make_graph(tesla_data, 'Tesla Stock Price Over Time')
```

```python

make_graph(gme_data, 'GameStop Stock Price Over Time')
```
